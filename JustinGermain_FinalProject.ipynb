## Introduction

"Legacy is a stupid thing!" Bill gates once shouted, "I don't want a legacy." To many it is strange to hear such an accomplished individual lambast an idea so grounded in historical truth. Whether they are scholarly books or a single paragraph in a database, biographies commonly act as the target of this existential panic. Before any analysis can occur, the .txt files must be imported into the Jupyter notebook. The following code reads in the four professions to be analyzed so that each biography will be an element in a list. Then, to allow for initial analysis and an accessible display, we will transform these texts into a pandas dataframe. Each profession will be a column and each row will contain the respective biography.

```python
athlete_path = '../fp/Updated/Profession/Athletes/'
author_path = '../fp/Updated/Profession/Author/'
politician_path = '../fp/Updated/Profession/Politician/'
scientist_path = '../fp/Updated/Profession/Scientist/'

athlete_files = os.listdir(athlete_path)
author_files = os.listdir(author_path)
politician_files = os.listdir(politician_path)
scientist_files = os.listdir(scientist_path)

athlete_texts = [open(athlete_path+file_name, encoding='utf-8').read() for file_name in athlete_files]
author_texts = [open(author_path+file_name, encoding='utf-8').read() for file_name in author_files]
politician_texts = [open(politician_path+file_name, encoding='utf-8').read() for file_name in politician_files]
scientist_texts = [open(scientist_path+file_name, encoding='utf-8').read() for file_name in scientist_files]

#This creates a dataframe with Athlete biographies, and then adds a column for each subsequent profession
df_profession = pandas.DataFrame(athlete_texts, columns = ['Athlete Biographies'])
df_profession['Author Biographies'] = pandas.DataFrame(author_texts)
df_profession['Politician Biographies'] = pandas.DataFrame(politician_texts)
df_profession['Scientist Biographies'] = pandas.DataFrame(scientist_texts)
df_profession
```

```python
#This takes each biography, separates them so that each word is an element in a list, and then counts and averages the number of elements
df_profession['Athlete Word Count'] = df_profession['Athlete Biographies'].str.split() 
df_profession['Athlete Word Count'] = df_profession['Athlete Word Count'].apply(len)
athlete_average_count = df_profession['Athlete Word Count'].mean()

df_profession['Author Word Count'] = df_profession['Author Biographies'].str.split() 
df_profession['Author Word Count'] = df_profession['Author Word Count'].apply(len)
author_average_count = df_profession['Author Word Count'].mean()

df_profession['Politician Word Count'] = df_profession['Politician Biographies'].str.split() 
df_profession['Politician Word Count'] = df_profession['Politician Word Count'].apply(len)
politician_average_count = df_profession['Politician Word Count'].mean()

df_profession['Scientist Word Count'] = df_profession['Scientist Biographies'].str.split() 
df_profession['Scientist Word Count'] = df_profession['Scientist Word Count'].apply(len)
scientist_average_count = df_profession['Scientist Word Count'].mean()

print ("Average Word Length of Athlete Biographies")
print (athlete_average_count)
print ()
print ("Average Word Length of Author Biographies")
print (author_average_count)
print ()
print ("Average Word Length of Politician Biographies")
print (politician_average_count)
print ()
print ("Average Word Length of Scientist Biographies")
print (scientist_average_count)
```

Now that we know how large each professions' biographies typically are, we should aim to find out the kinds of words used within these biographies. To do this, we will create a Document Term Matrix in order to help us understand this data in an accessible format. NLTK could also be used to perform these operations, but we are using a DTM in order to explore the technique further and see what other possible operations it can conduct. This same principles extends to other techniques used in this exploratory analysis. While using dataframes, DTMs, or other structures may not be necessary to the technique, it helps to both understand and visualize the data in different ways. 

First, we will import the necessary packages. Then, the following code pre-processes the biographies by lowercasing the texts of each biography, removing all numbers, splitting the text on spaces, and removing punctuation. I have decided to leave stopwords in because I am curious about how often they appear across different professions. Finally, we will transform the cleaned biographies into Document Term Matrices expressed in pandas dataframes.

```python
from sklearn.feature_extraction.text import CountVectorizer
countvec = CountVectorizer()
import nltk
punctuation = list(string.punctuation)

#Cleans each column in the dataframe to allow for a better frequency count analysis
df_profession['Athlete_Clean'] = df_profession['Athlete Biographies'].str.lower()
df_profession['Athlete_Clean'] = df_profession['Athlete_Clean'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
df_profession['Athlete_Clean'] = df_profession['Athlete_Clean'].str.split()
df_profession['Athlete_Clean'] = [word for word in df_profession['Athlete_Clean'] if word not in punctuation]

df_profession['Author_Clean'] = df_profession['Author Biographies'].str.lower()
df_profession['Author_Clean'] = df_profession['Author_Clean'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
df_profession['Author_Clean'] = df_profession['Author_Clean'].str.split()
df_profession['Author_Clean'] = [word for word in df_profession['Author_Clean'] if word not in punctuation]

df_profession['Politician_Clean'] = df_profession['Politician Biographies'].str.lower()
df_profession['Politician_Clean'] = df_profession['Politician_Clean'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
df_profession['Politician_Clean'] = df_profession['Politician_Clean'].str.split()
df_profession['Politician_Clean'] = [word for word in df_profession['Politician_Clean'] if word not in punctuation]

df_profession['Scientist_Clean'] = df_profession['Scientist Biographies'].str.lower()
df_profession['Scientist_Clean'] = df_profession['Scientist_Clean'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
df_profession['Scientist_Clean'] = df_profession['Scientist_Clean'].str.split()
df_profession['Scientist_Clean'] = [word for word in df_profession['Scientist_Clean'] if word not in punctuation]

profession_dtm_athlete = pandas.DataFrame(countvec.fit_transform(df_profession['Athlete_Clean'].apply(lambda x: ' '.join(x))).toarray(), columns=countvec.get_feature_names(), index = df_profession.index)
profession_dtm_author = pandas.DataFrame(countvec.fit_transform(df_profession['Author_Clean'].apply(lambda x: ' '.join(x))).toarray(), columns=countvec.get_feature_names(), index = df_profession.index)
profession_dtm_politician = pandas.DataFrame(countvec.fit_transform(df_profession['Politician_Clean'].apply(lambda x: ' '.join(x))).toarray(), columns=countvec.get_feature_names(), index = df_profession.index)
profession_dtm_scientist = pandas.DataFrame(countvec.fit_transform(df_profession['Scientist_Clean'].apply(lambda x: ' '.join(x))).toarray(), columns=countvec.get_feature_names(), index = df_profession.index)
```

The main analysis that we can undertake using these DTM's is frequency counting. Through this, we can understand the most frequent words used in each biography. The following code sorts each word in descending order by the frequency at which it is used in the biographies, then displays the top 50 most frequent words for each profession.

```python
print ("Top 50 Words Used in Athletes' Biographies")
print ()
print (profession_dtm_athlete.sum().sort_values(ascending=False)[:50])
print ()
print ("Top 50 Words Used in Authors' Biographies")
print ()
print (profession_dtm_author.sum().sort_values(ascending=False)[:50])
print ()
print ("Top 50 Words Used in Politicians' Biographies")
print ()
print (profession_dtm_politician.sum().sort_values(ascending=False)[:50])
print ()
print ("Top 50 Words Used in Scientists' Biographies")
print ()
print (profession_dtm_scientist.sum().sort_values(ascending=False)[:50])
```

These frequency counts have intriguing results in terms of biographical themes.

**Athletes**

Biographies of athletes tend to use words associated with times and names more frequently than other professions. This makes sense, as athletes are often only prominent during a specific time period. Common words like "first," "life," and "after," reveal how athletes are memorialized in terms of their influence within a few decades of their life. In addition, since many become household names and are often ranked along side other athletes, it is understandable why names can be more prominent than words generally associated with sports. In this case, although the names appearing on the list of most frequent words are names of individuals with selected biographies, it is still impressive that they are more frequent than sports-related terms like "game" and "play." Such a large focus on names and times suggest that biographies of athletes primarily focus on individuals' relationship to the sport they play when describing them as a person.

**Authors**

The main theme that appears in the biographies of authors is the importance of their written works. Commonly-appearing words like "published," "novel," and "book" suggest that one of the best ways to refer to authors is through their writings. Unlike athletes, who are described in relation to their respective time periods, authors are associated mostly with the products that they create, relatively irrespective of when they were created. The only exception to this is the prominence of the word "first," yet this may be used to reference the order in which authors' works were written. While athletes are described based on their relationship to the athletic world, authors are described through what they produced for the world.

**Politicians**

Words like "president," "republican," and "state," reveal that it is necessary to examine the professional roles of politicians in order to properly describe who they were as individuals. Just as names and times located athletes in their specific period of accomplishments, these affiliations locate politicians in their positions. Furthermore, words like "national" and "city" illuminate how important scale is when describing the life of a politician. This scale delimits the area that their decisions affected. Therefore, politicians are most commonly described through their associations to societal institutions and bodies.

**Scientists**

Similar to politicians, scientists are commonly framed in relation to the scale and bodies in which they work. The prominent inclusion of "university," "national," and "laboratory" provide important context in understanding where and how scientists act. Words like "research," "physics," and "medical" also serve to isolate scientists' accomplishments within a specific field. Both athletes and scientists are commonly described through the word "first." This is understandable, since both are often noted for being the first in their field to complete a certain action. On the other hand, politicians and authors are not usually known for this. With this information in hand, biographers seem to describe scientists in terms of both their work and the scope in which that work is conducted. 

Surprisingly, all four professions have a similar frequency of stopwords, with the top 18-21 most frequent words being stopwords. This confirms that biographies have generally similar structures that change little across different professions. 

Our final technique of this lexical selection analysis will be to find tf-idf weights for each of the professions' biographies. Unlike simply finding the most frequent words within each career field, this will uncover the most distinct words. In other words, tf-idf weights help to find the words that appear frequently but not across many documents. This will be especially helpful when analyzing differences within each profession, rather than comparing solely across professions.

First we must import the necessary packages. Then we will create another Document Term Matrix expressed through a pandas dataframe. This time, however, the value that appears in each cell represents tf-idf scores instead of the number of times that a word appears within the biographies of each profession. The following code calculates and displays the top twenty words with the highest tf-idf scores for each professions' biographies. 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfvec = TfidfVectorizer()

dtm_tfidf_athlete = pandas.DataFrame(tfidfvec.fit_transform(df_profession['Athlete_Clean'].apply(lambda x: ' '.join(x))).toarray(), columns=tfidfvec.get_feature_names(), index = df_profession.index)
dtm_tfidf_author = pandas.DataFrame(tfidfvec.fit_transform(df_profession['Author_Clean'].apply(lambda x: ' '.join(x))).toarray(), columns=tfidfvec.get_feature_names(), index = df_profession.index)
dtm_tfidf_politician = pandas.DataFrame(tfidfvec.fit_transform(df_profession['Politician_Clean'].apply(lambda x: ' '.join(x))).toarray(), columns=tfidfvec.get_feature_names(), index = df_profession.index)
dtm_tfidf_scientist = pandas.DataFrame(tfidfvec.fit_transform(df_profession['Scientist_Clean'].apply(lambda x: ' '.join(x))).toarray(), columns=tfidfvec.get_feature_names(), index = df_profession.index)

print ("Top 20 Most Distinctive Words in Athletes' Biographies")
print (dtm_tfidf_athlete.max().sort_values(ascending=False)[:20])
print ()
print ("Top 20 Most Distinctive Words in Authors' Biographies")
print (dtm_tfidf_author.max().sort_values(ascending=False)[:20])
print ()
print ("Top 20 Most Distinctive Words in Politicians' Biographies")
print (dtm_tfidf_politician.max().sort_values(ascending=False)[:20])
print ()
print ("Top 20 Most Distinctive Words in Scientists' Biographies")
print (dtm_tfidf_scientist.max().sort_values(ascending=False)[:20])
```

These results indicate that my original hypothesis is correct. The most distinct words are those of the biographical subjects, since their names disproportionately occur within their own biographies. Yet these weights do reveal some other interesting results. Words that are expected to be common across all individuals' biographies seem to be more distinct than assumed. For instance, "computer" is quite distinct within scientists' biographies. Furthermore, feminine pronouns like "she" and "her" are very distinct across all professions. This makes sense, as most of the randomly selected biographies are focused on men.

## 2. Topic Modeling

We can use topic modeling to learn more about biographies as a whole, rather than just by individuals' professions. This technique will help to discover which identities are more salient than others when writing biographies. How often do social identities emerge as a pervasive theme across biographies? Are there other, non-social factors that are common to all biographies? Answering these questions is important because it allows for a better understanding of what biographies share, even when focusing on vastly different individuals. Discovering these common threads is vital to understanding how biographies are created in the first place. Perhaps individuals are primarily described by the jobs that they do, or perhaps by their race or nationality. Of course a larger corpus would allow for this understanding to be more representative and increase the possible themes that may emerge. Yet for this analysis, I hypothesize that a person's profession will be most prominent within the generated topics. Additionally, I believe that racial identifications will be more salient than national associations.

Topic modeling examines each word in the set of biographies and determines the words surrounding it. By doing this for each word, the algorithm is able to quantify which words appear in similar contexts as other words. This allows multiple topics to be generated and displayed. 

We will use the dataframe of all of our biographies for this analysis. We will begin the topic modeling process by importing the necessary packages and setting the preliminary conditions. I have chosen to create six topics in order to better understand which identities are the most prominent compared to others, knowing that there are more than six subcategories. 

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

#Selects how many topics to create and the number of top words shown within each 
n_samples = 2000
n_topics = 6
n_top_words = 15

def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
    print()
    
tf_vectorizer = CountVectorizer(max_df=200, min_df=10,
                                max_features=None,
                                stop_words='english'
                                )

tf = tf_vectorizer.fit_transform(df_all.text)

print("Fitting LDA models with tf features, "
      "n_samples=%d and n_topics=%d..."
      % (n_samples, n_topics))

lda = LatentDirichletAllocation(n_topics=n_topics, max_iter=20,
                                learning_method='online',
                                learning_offset=80.,
                                total_samples=n_samples,
                                random_state=0)
lda.fit(tf)

print("\n Six Topics in LDA model for biographies:")
tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
```

These topics illuminate important categorization information regarding the role of identities in biographies. First, out of the six topics, two (Topic 4 and 5) are related to athletics. Surprisingly, only 24 out of the 160 biographies (15%) were about athletes (including the set of twenty for the analysis of professions and four from biographies defined by race or nationality). This indicates that there may be a large distinction within the words used to describe athletes in their biographies. The differences between Topics 4 and 5 can help to shed more light on this possible divergence. Topic 4 talks heavily about competition and time. Words like "champion," "title," and "individual" create connotations of athletic endeavors as intense matches between bitter rivals. Topic 5, however, is quite different. With prominent words like "team," "season," "runs," and "player," this topic deals primarily with the game structure and teamwork involved in different sports. Rather than competition or end results, this topic emphasizes the actual act of playing a sport. It is likely that biographers framed athletes quite distinctly along these lines, therefore creating enough difference to create two separate topics.

Other profession-based topics also appear in this topic modeling analysis. With words like "book," "fiction," and "literature," Topic 0 focuses on the act of writing. While this is likely prominent due to the high occurrence of authors (31 in total), many individuals with biographies have written books themselves. This extends across professions, as famous athletes, politicians, and musicians have attempted to create both fictional and nonfictional works. Writing, therefore, holds an important place in biographies. Just as biographies chronicle an individual's life, so do their writings. Topic 2 is also defined through professional identification, yet this time in a career field not included in the lexical selection analysis. Such a focus on music and performance stems from the fact that 32 of the examined biographies are of actors, dancers, musicians, or artists. 

Topics 1 and 3 are defined in non-professional ways, with the former centered around a distinctly Asian-American identity. It is interesting to see such a strong association with an Asian heritage, while other racial or national associations examined (English, Mexican, African-American) do not appear strongly within any topic. This may suggest that, at least out of these texts, individuals with an Asian-American heritage are described more along identity-based lines than other races or nationalities in their biographies. It is important to note, however, that this is not representative of all possible biographies. It is best representative of biographies present in online databases.

Instead of focusing on race, nationality, or profession, Topic 3 seems to be concerned with institutions associated with elites. Terms connected to academics, scholarly research, and politics appear heavily within this topic. Therefore, this may imply that the biographies examined in this topic modeling analysis may surround people who were involved within elite circles. This makes logical sense; biographers are likely to chronicle the lives of important, famous, or widely successful people. Yet this shows that studying biographies may not illuminate much information about the lives of most people not seemingly successful enough to be written about. Studying biographies essentially means studying elite society.

In order to build off of the above analyses, we will now run a second topic modeling algorithm on the same corpus. This time, however, we will only create three topics and see what differences occur.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

#Here we have changed the number of topics from 6 to 3
n_samples = 2000
n_topics = 3
n_top_words = 15

def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
    print()
    
tf_vectorizer = CountVectorizer(max_df=200, min_df=10,
                                max_features=None,
                                stop_words='english'
                                )

tf = tf_vectorizer.fit_transform(df_all.text)

print("Fitting LDA models with tf features, "
      "n_samples=%d and n_topics=%d..."
      % (n_samples, n_topics))

lda = LatentDirichletAllocation(n_topics=n_topics, max_iter=20,
                                learning_method='online',
                                learning_offset=80.,
                                total_samples=n_samples,
                                random_state=0)
lda.fit(tf)

print("\nThree Topics in LDA model for biographies:")
tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
```

With only three topics, we still have one topic dedicated to athletics, teamwork, and competition. While words like "film," and "music," appear, Topic 2 seems to combine the two athletics topics of our above analysis. Instead of Topic 4's focus on rivalry and competition, however, this new athletic topic is much more similar to Topic 5 in the previous analysis. Through this examination, athletics is defined as a collective action. Therefore, biographers may frame sports more distinctly as an act of group-belonging rather than an act of competition.

While our previous topic modeling had multiple topics revolving around professions, this model only has one. Topics 0 and 1 seem to revolve primarily around research, work, and politics. While the latter may be considered to be associated with a professional identity, as one of our studied career fields included politicians, it could also imply that many individuals being written about were simply involved in political discourse. ## 3. Vector-Space Modeling

Our previous analyses have examined the lexical structure and content of biographies in order to determine how social factors influence their themes. By using vector-space models, we can take all of these into account in a single technique. Calculating the euclidean distances between different forms of biographies can draw out how similar biographical categories are to each other. This technique uses DTM frequency counts in order to directly compare the words used of biographies within one identity to another. For example, how similar are biographies of politicians to biographies of Asian-Americans? Previous techniques have only hinted at this direct comparison. We used lexical selection and topic modeling to discover key words and themes that defined each kind of biography, yet vector-space modelling gives more weight to the whole of each biographical category.

To conduct such an analysis, I sorted each of my 160 biographies into their respective subcategories. Next, I put all twenty biographies of the same subcategory into a single .txt document. The code below imports all necessary packages and creates a list with each biographical subcategory as an element. It then creates a DTM of each element (which expresses each subcategory as a vector) and then calculates the euclidean distance between each. The DTM is the most necessary part of this analysis, since it transforms qualitative data into quantitative, vector-based data. Without doing this, it would be much more difficult to compare these biographies in this manner.

This code visualizes the distances between each subcategory by plotting the values on a two-dimensional plane. The distances between two points represents their euclidean distances. While this visualization provides a helpful, top-down picture of how each subcategory compares to others, it is difficult to understand which points are closest to each other. Therefore, I create a second visualization that forms a cluster tree through Ward's method. It groups subcategories hierarchically based on their euclidean distances. This technique informs us which biographies are most similar to each other, thereby providing insight into how people of different identities are described.

I use two different corpora for this analysis. The first model is based off of all 160 biographies. Second, I limit my corpus to just those based on professions in order to compare results from this analysis to the previous lexical selection technique. Since the topic modeling analysis showed that there were few racial or national topics, I believe it is unlikely that racial and national subcategories will be consistently near each other in terms of euclidean distances. Regarding profession-based biographies, I hypothesize that athletes will stand out compared to the other three. Since athletes commanded two topics in the previous analysis, I believe that their similarities with other biographies will be less apparent.

```python
import numpy as np 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import euclidean_distances
import matplotlib.pyplot as plt
from sklearn.manifold import MDS

filenames = ['../fp/Updated/Vector/AthletesAll.txt',
            '../fp/Updated/Vector/AuthorsAll.txt',
            '../fp/Updated/Vector/PoliticiansAll.txt',
            '../fp/Updated/Vector/ScientistsAll.txt', 
            '../fp/Updated/Vector/AfricanAmericansAll.txt',
            '../fp/Updated/Vector/AsianAmericansAll.txt.',
            '../fp/Updated/Vector/EnglishAll.txt',
            '../fp/Updated/Vector/MexicansAll.txt']

vectorizer = CountVectorizer(input='filename', encoding='utf-8',stop_words='english') 
dtm = vectorizer.fit_transform(filenames)  
vocab = vectorizer.get_feature_names()  

#Creates a dtm for each element in the list
dtm = dtm.toarray() 
vocab = np.array(vocab)

#Sets variable equal to the euclidean distance between the elements in the list
euc_dist = euclidean_distances(dtm)

mds = MDS(n_components=2, dissimilarity="precomputed", random_state=1)

pos = mds.fit_transform(euc_dist) 
xs, ys = pos[:, 0], pos[:, 1]

#Changes the name on the plot to be more understandable visually
names = [os.path.basename(fn).replace('All.txt', '') for fn in filenames]

for x, y, name in zip(xs, ys, names):
    plt.scatter(x, y)
    plt.text(x, y, name)

plt.show()
```

```python
from scipy.cluster.hierarchy import ward, dendrogram
linkage_matrix = ward(euc_dist)

#Creates a branched tree dendrogram that hierarchically clusters the biographies
dendrogram(linkage_matrix, orientation="right", labels=names)

plt.tight_layout()

plt.show()
```

```python
#Applies the same technique, but to just professions rather than all biographies
filenames = ['../fp/Updated/Vector/AthletesAll.txt',
            '../fp/Updated/Vector/AuthorsAll.txt',
            '../fp/Updated/Vector/PoliticiansAll.txt',
            '../fp/Updated/Vector/ScientistsAll.txt', ]

vectorizer = CountVectorizer(input='filename', encoding='utf-8',stop_words='english') 
dtm = vectorizer.fit_transform(filenames)  
vocab = vectorizer.get_feature_names()  

dtm = dtm.toarray() 
vocab = np.array(vocab)

from sklearn.metrics.pairwise import euclidean_distances
euc_dist = euclidean_distances(dtm)

mds = MDS(n_components=2, dissimilarity="precomputed", random_state=1)

pos = mds.fit_transform(euc_dist)  
xs, ys = pos[:, 0], pos[:, 1]

names = [os.path.basename(fn).replace('All.txt', '') for fn in filenames]


for x, y, name in zip(xs, ys, names):
    plt.scatter(x, y)
    plt.text(x, y, name)

plt.show()
```
